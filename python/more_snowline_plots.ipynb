```python
import numpy as np
from matplotlib import pyplot as plt
import C_to_O as CO
from T_freeze import T_freeze, Rdes, tevap
from C_to_O import T_freeze_H20, T_freeze_CO2, T_freeze_CO
import matplotlib.cm as cm

from drift_timescales import tr, Tdisk, ts, lambdamfp, cdisk, tdes, rf, r_freeze, read_drag_file, read_drag_file_many, r_stop
from scipy.optimize import brentq
from utils.constants import cmperau
```

```python
s, a, arrayH2O = read_drag_file_many('s_neg3_8.txt', 'H2O', 50, 50)
s, a, arrayCO2 = read_drag_file_many('s_neg3_8.txt', 'CO2', 50, 50)
s, a, arrayCO = read_drag_file_many('s_neg3_8.txt', 'CO', 50, 50)
```

```python
rfreeze_H2O = r_freeze(18., 5800., 1e6)
rfreeze_CO2 = r_freeze(44., 2000., 3e5)
rfreeze_CO = r_freeze(28., 850., 1e6)
```

```python
f, (ax1, ax2, ax3) = plt.subplots(3, sharex = True, figsize = (3.5, 8))

rnb = linspace(0, 256, len(s))

for i in range(len(s)):
    ax1.loglog(a, arrayH2O[i], c = cm.rainbow(int(rnb[i])))
    ax2.loglog(a, arrayCO2[i], c = cm.rainbow(int(rnb[i])))
    ax3.loglog(a, arrayCO[i], c = cm.rainbow(int(rnb[i])))
ax3.set_xlim(xmin = 0.3)
ax3.set_ylim(ymin = 0.3)
ax1.set_ylabel('Desorption distance [AU]')
ax2.set_ylabel('Desorption distance [AU]')
ax3.set_ylabel('Desorption distance [AU]')
ax3.set_xlabel('Initial distance [AU]')

ax1.axhline(y = rfreeze_H2O, ls = '--', c = 'black')
ax2.axhline(y = rfreeze_CO2, ls = '--', c = 'black')
ax3.axhline(y = rfreeze_CO, ls = '--', c = 'black')

ax1.axvline(x = rfreeze_H2O, ls = '--', c = 'black')
ax2.axvline(x = rfreeze_CO2, ls = '--', c = 'black')
ax3.axvline(x = rfreeze_CO, ls = '--', c = 'black')

ax1.text(0.7, 55, 'H2O')
ax2.text(0.7, 55, 'CO2')
ax3.text(0.7, 55, 'CO')

ax1.set_ylim(ymin = 0.3, ymax = 1e2)
ax2.set_ylim(ymin = 0.3, ymax = 1e2)
ax3.set_ylim(ymin = 0.3, ymax = 1e2)

plt.tight_layout()
f.subplots_adjust(hspace = 0)

#plt.savefig('desorption_distace_many.pdf')
    
```
