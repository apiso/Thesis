```python
gas_sol_ratio1, gas_sol_ratio2, gas_sol_ratio3 = [], [], []
fH2O = 1e-4

for i in range(len(r) - 1):
    gas_sol_ratio1 = np.append(gas_sol_ratio1, (dMgas_dt(r[i], r[i + 1], 0.01, betaS = 1.) + Mgas_dt(r[i], r[i + 1], 1., 44, 5800, betaS = 1.)) \
        / (dMsol_dt(r[i], r[i + 1], 1., 44., 5800, betaS = 1.) + Msol_dt(r[i], r[i + 1], 1., 44, 5800, betaS = 1.)))
    gas_sol_ratio2 = np.append(gas_sol_ratio2, (dMgas_dt(r[i], r[i + 1], 0.01, betaS = 1.) + Mgas_dt(r[i], r[i + 1], 1., 44, 5800, betaS = 1.)) \
        / (dMsol_dt(r[i], r[i + 1], 1., 44., 5800, betaS = 1.) + Msol_dt(r[i], r[i + 1], 1e5, 44, 5800, betaS = 1.)))
    gas_sol_ratio3 = np.append(gas_sol_ratio3, (dMgas_dt(r[i], r[i + 1], 0.01, betaS = 1.) + Mgas_dt(r[i], r[i + 1], 1., 44, 5800, betaS = 1.)) \
        / (dMsol_dt(r[i], r[i + 1], 1., 44., 5800, betaS = 1.) + Msol_dt(r[i], r[i + 1], 1e-5, 44, 5800, betaS = 1.)))
```

```python
f, (ax1, ax2, ax3) = plt.subplots(3, sharex = True, figsize = (3.5, 7))

ax1.semilogx(r[:-1], fH2O * gas_sol_ratio1, label = r's=1 cm')
ax2.semilogx(r[:-1], fH2O * gas_sol_ratio2, label = r's=$10^5$ cm')
ax3.semilogx(r[:-1], fH2O * gas_sol_ratio3, label = r's=$10^{-5}$ cm')

ax3.set_xlabel('r [AU]')
ax1.set_ylabel(r'$\Delta M_{H_2O,g}/\Delta M_{H_2O,s}$')
ax2.set_ylabel(r'$\Delta M_{H_2O,g}/\Delta M_{H_2O,s}$')
ax3.set_ylabel(r'$\Delta M_{H_2O,g}/\Delta M_{H_2O,s}$')

ax1.legend(frameon = False)
ax2.legend(frameon = False)
ax3.legend(frameon = False)

plt.tight_layout()
#f.subplots_adjust(hspace = 0)

#plt.savefig('../figs/H2O_gas_to_dust.pdf')
```
